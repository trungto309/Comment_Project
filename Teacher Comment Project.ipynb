```python
def head_students(file):
    """This funtion takes in a csv file name
    (with the extension) and produces a
    tuple with the first value being the
    header of the csv file (the first line
    stored as a list) and a second value of
    the rest of the values (stored as a
    list)."""
    import csv
    student_info = []
    with open(file) as f:
        data = csv.reader(f, delimiter=',')
        for index, line in enumerate(data):
            if index == 0:
                header = line
            else:
                student_info.append(line)
    return header, student_info


def first(quiz, problem, final, name, pronouns, grade):
    import random as rand
    """This function takes in the following values in the exact order:
    percentage stores a whole number representing the student's quiz average
    (eg. 86% would be 86), percentage as whole number representing student's
    problem set percentage average, percentage as whole number representing
    student's fianl exam percentage, a string of the student's first name, and
    a list of their pronouns (following the EXACT order:
    [subjective,objective,possesive]-[she,her,hers] for feminine pronouns).
    The function will take these inputs and produce the first paragraph of the
    student's comment, returned as a string."""
    name = name.title()
    # The code below deals with the dilemma of feminine pronouns having "her"
    # AND 'hers' as possesive. For example, you would say "his shoes,"
    # but you would not say "hers shoes." Thus, you must say, "her shoes."
    if 'she' in pronouns:
        num = 1
    else:
        num = -1
    # The if/else below deals with people who use they pronouns
    # it turns "was" into "were" and "is" to "are"
    if 'they' in pronouns:
        was = "were"
        myis = 'are'
    else:
        was = 'was'
        myis = 'is'
    # The student's work is deemed satisfactory if they received an A or a B
    # Otherwise, they encountered difficulty
    # Below are 4 similar responses for the beginning of the comment
    # if the students grade was good
    good_grade1 = f"I am pleased with {name.title()}'s"
    good_grade1 += " progress this semester."
    good_grade2 = f"I believe {name.title()} put in"
    good_grade2 += " good effort this semester."
    good_grade3 = f"{name.title()} has made great"
    good_grade3 += " strides this semester."
    good_grade4 = f"{name.title()} has continued to push"
    good_grade4 += f" {pronouns[1]}self this semester."
    # A list of the good responses is then generated to produce
    # a list which can produce "random" responses
    # This decreases the chance the student will know that the comment
    # is generated by a program
    good_grade = [good_grade1,
                  good_grade2,
                  good_grade3,
                  good_grade4]
    # The same random list is generated for "bad" grades
    bad_grade1 = f"{name.title()} has encountered difficulty this semester"
    bad_grade2 = f"{name.title()} has encountered"
    bad_grade2 += " some difficulties this semester"
    bad_grade3 = f"{name.title()} has experienced some"
    bad_grade3 += " difficulties this semester"
    bad_grade4 = f"{name.title()} has experienced difficulty this semester"
    bad_grade = [bad_grade1,
                 bad_grade2,
                 bad_grade3,
                 bad_grade4]
    # If the student's grade was good, they are assigned a
    # random response in the good responses list
    if grade > 85:
        sentence1 = good_grade[rand.randint(0, len(good_grade)-1)]
    # If the student's grade was not good, they are assigned a
    # random response in the bad responses list
    else:
        sentence1 = bad_grade[rand.randint(0, len(bad_grade)-1)]
        sentence1 += f" but {pronouns[0]} {myis} persevering."
    # Since this code is not going to change
    # no matter the student's performance,
    # it can be added out of the if/else statement.
    sentence2 = f" {pronouns[num].title()} problem set average was {problem}%,"
    # For the scope of this project, the value of success is 85% and above. If
    # there was more time available for the project, there would be more minute
    # success judgements(0-75=needs a lot of work,75-80=needs work,80-85=okay,
    # 85-90=good,90-100=amazing).
    # The same random response list is generated for the problem set average
    good_problem1 = f" which indicates that {pronouns[0]} had a solid"
    good_problem1 += " understanding of the majority of the topics"
    good_problem1 += f" covered this semester and that {pronouns[0]}"
    good_problem1 += f" {was} consistent about meeting deadlines."
    good_problem2 = f" which shows that {name} put in much effort and"
    good_problem2 += " had a solid grasp of the topics covered this"
    good_problem2 += " semeseter."
    good_problem3 = f" {pronouns[0].title()} has a good understanding of the"
    good_problem3 += " majority of the topics covered this semester"
    good_problem3 += f" and that {pronouns[0]} {was} consistent about"
    good_problem3 += " meeting deadlines."
    good_problem4 = f" which tells me that {name} really had"
    good_problem4 += " a solid understanding of the majority of topics"
    good_problem4 += f" covered this semester and that {pronouns[0]} {was}"
    good_problem4 += f" consistent about meeting deadlines."
    good_problem = [good_problem1,
                    good_problem2,
                    good_problem3,
                    good_problem4]
    bad_problem1 = f" which indicates that {pronouns[0]}"
    bad_problem1 += f" had gaps in {pronouns[num]} understanding and {was}"
    bad_problem1 += " inconsistent about meeting deadlines."
    bad_problem2 = f" which shows me that {pronouns[0]}"
    bad_problem2 += f" had gaps in {pronouns[num]} understanding"
    bad_problem2 += " and did not consistently meet deadlines."
    bad_problem3 = f" which shows that {name.title()} had"
    bad_problem3 += f" several gaps in {pronouns[num]}"
    bad_problem3 += f" understanding and {was} not very consitent"
    bad_problem3 += " about meeting deadlines."
    bad_problem = [bad_problem1,
                   bad_problem2,
                   bad_problem3]
    if problem >= 85:
        sentence2 += good_problem[rand.randint(0, len(good_problem)-1)]
    else:
        sentence2 += bad_problem[rand.randint(0, len(bad_problem)-1)]
    # Since this code is not going to change no matter
    # the student's performance, it can be added out of
    # the if/else statement.
    # The same random list is generated for the quiz average
    # and final exam score
    sentence3 = f"{pronouns[num].title()} quiz average and"
    sentence3 += f" final exam score were {quiz}% and {final}%, respectively, "
    good_quizfinal1 = f" which demonstrated that {pronouns[0]}"
    good_quizfinal1 += " grasped the new content thoroughly and"
    good_quizfinal1 += " could hand write algorithms without"
    good_quizfinal1 += " using outside resources."
    good_quizfinal2 = f" which shows that {pronouns[0]} understood"
    good_quizfinal2 += " the new content thoroughly and could easily"
    good_quizfinal2 += " hand write programs without outside resourses."
    good_quizfinal = [good_quizfinal1,
                       good_quizfinal2]
    bad_quizfinal1 = f" which demonstrated that {pronouns[0]}"
    bad_quizfinal1 += " did not grasp the new content thoroughly"
    bad_quizfinal1 += " and could not hand write algorithms without"
    bad_quizfinal1 += " using outside resources."
    bad_quizfinal2 = f" which tells me that {pronouns[0]} did not"
    bad_quizfinal2 += " grasp the new material often and could not"
    bad_quizfinal2 += " hand programs without using outside resources."
    bad_quizfinal3 = f" which shows me that {pronouns[0]} did not"
    bad_quizfinal3 += " really grasp the new content and couldn't"
    bad_quizfinal3 += " had write algorithms without"
    bad_quizfinal3 += " using outside resources."
    bad_quizfinal = [bad_quizfinal1,
                     bad_quizfinal2,
                     bad_quizfinal3]
    if (quiz+final)/2 >= 85:
        sentence3 += good_quizfinal[rand.randint(0, len(good_quizfinal)-1)]
    else:
        sentence3 += bad_quizfinal[rand.randint(0, len(bad_quizfinal)-1)]
    # Since three seperate strings were created, they have to be
    # joined together as a list to be returned as one string
    sentences = [sentence1, sentence2, sentence3]
    return ' '.join(sentences)


def second(bank, name, pronouns):
    """This function takes the folling variables in the following order:
    student's bank percentage score represented as a whole number without the %
    sign, the student's first name as a string, and and a list of their
    pronouns (following the EXACT order:
    [subjective,objective,possesive]-[she,her,hers] for feminine pronouns).
    It uses these inputs to produce a string of their second paragraph."""
    # The code below deals with the 'a'/'an' error. If the student's bank grade
    # begins with an 8, thus beginning with a vowel, then it will produce 'an'.
    # Otherwise it will produce 'a'
    if list(str(bank))[0] == '8':
        a = 'an'
    else:
        a = 'a'
    # Dealing with people who use "they" pronouns
    # This switches "is" to "are" and "was" to "were"
    if 'they' in pronouns:
        was = "were"
        myis = 'are'
    else:
        was = 'was'
        myis = 'is'
    # The code below deals with the dilemma of feminine pronouns having
    # "her" AND 'hers' as possesive. For example, you would say "his
    # shoes," but you would not say "hers shoes."
    # Thus, you must say, "her shoes."
    if 'she' in pronouns:
        num = 1
    else:
        num = -1
    # Since this code is not going to change no matter the
    # student's performance, it can be added out of the if/else statement.
    sentence1 = f"{pronouns[num].title()} group Bank Project grade"
    sentence1 += f" was {a} {bank}%,"
    if bank >= 85:
        sentence1 += f" which indicated that {pronouns[-1]} effort,"
        sentence1 += " time management, and collaboration relating"
        sentence1 += " to larger group assignments was good."
    else:
        sentence1 += f" which indicated that {pronouns[-1]} effort,"
        sentence1 += " time management, and collaboration relating"
        sentence1 += " to larger group assignments was unsatisfactory."
    # Since this code is not going to change no matter
    # the student's performance, it can be added out of the if/else statement.
    sentence1 += f' In addition, '
    if bank >= 85:
        sentence1 += f"{pronouns[0]} demonstrated significant enthusiasm and"
        sentence1 += " dedication when we participated in the Hackathons. "
    else:
        sentence1 += f"{pronouns[0]} tended to lose focus when"
        sentence1 += " we participated in the Hackathons. "
    sentence1 += f"These assignments allowed students to"
    sentence1 += " put forth as much effort as they wished to, "
    if bank >= 85:
        sentence1 += f"and {pronouns[0]} chose to use these opportunities"
        sentence1 += f" to seriously challenge {pronouns[-1]}self, think"
        sentence1 += f" creatively, and advance {pronouns[num]} skills."
    else:
        sentence1 += f"and {pronouns[0]} often lost focus instead of working"
        sentence1 += f" on advancing {pronouns[num]} skills. "
    # Since one string was updated throughout the function,
    # only the single string needs to be returned
    return sentence1


def third(daily, stackoverflow, extrahelp, grade, pronouns, name):
    # Dealing with the previously define "she" pronoun issue
    if 'she' in pronouns:
        num = 1
    else:
        num = -1
    # The code below checks whether or not the student
    # used StackOverflow and defines a proper sentence
    if stackoverflow.lower().strip() == "yes":
        sentence_3 = f"In regard to using available resources like"
        sentence_3 += f" StackOverflow, {name.title()} wisely utilized"
        sentence_3 += " these resources to troubleshoot and gain inspiration"
        sentence_3 += " and internalized the solutions that"
        sentence_3 += f" {pronouns[0]} found."
    else:
        sentence_3 = f"In regard to using available resources like"
        sentence_3 += f" StackOverflow, {name.title()} sometimes"
        sentence_3 += " used these resources without fully understanding"
        sentence_3 += " what the code did."
    # The code below check if the student
    # asked for extra help and defines a proper sentence
    if extrahelp.lower().strip() == 'no':
        last_sentence = f' {pronouns[0].title()} rarely if ever stopped by'
        last_sentence += f' for extra help, which I think would'
        last_sentence += f' have really helped {pronouns[1]} to'
        last_sentence += f' solidify {pronouns[num]} skills.'
    else:
        last_sentence = f'In terms of coming in for extra help,'
        last_sentence += f' {pronouns[0].title()} rarely if ever'
        last_sentence += f' needed to do so because {pronouns[0]}'
        last_sentence += ' {was} so easily able to pick up the'
        last_sentence += ' concepts in class.'
    # The code below checks if the student worked
    # alone or with friends during class and defines
    # a proper response according to their grade
    if grade >= 85 and daily.lower().strip() == 'no':
        first_sentence = f'On the daily coding activities,'
        first_sentence += f' {pronouns[0]} worked mostly by'
        first_sentence += f' {pronouns[1]}self since the material'
        first_sentence += f' came easily to {pronouns[1]}, but {pronouns[0]}'
        first_sentence += ' {was} happy to help a classmate'
        first_sentence += ' when they got stuck.'
    else:
        first_sentence = f'On the daily coding activities, it would have'
        first_sentence += f' been beneficial for {pronouns[1]} to work'
        first_sentence += f' with {pronouns[-1]} classmates rather rather'
        first_sentence += ' than working alone.'
    sentences_1 = [sentence_3, last_sentence, first_sentence]
    return ' '.join(sentences_1)


def finalgrade(final, problem, bank, quiz, name, pronouns):
    # Dealing with the previously define "she" pronoun problem
    if 'she' in pronouns:
        num = 1
    else:
        num = -1
    final = final*0.4
    # the quiz grade combined with final exam grade(final was worth 2 quizzes)
    problem = (((problem*7)+(bank*2))/9)*0.4
    quiz = quiz*0.2
    # Each category of grades and its weighting
    finale = final+problem+quiz
    # The code below tells the student, if they do not
    # have an A in the class already, the score they would
    # have had to receive on the final in order to receive an
    # A in the class. In some cases it might be unrealistic
    # (if the student recieve a zero on everything else, for
    # example), but we thought we would like to know it as students
    if finale < 89.5:
        i = final
        # Because the range function cannot
        # iterate by decimals, the numbers have to multiplied
        # by a thousand to create a massive whole number
        a = int(round((89.5)*1000))
        problem = int(round(problem*1000))
        quiz = int(round(quiz*1000))
        while int(round(i*1000))+problem+quiz != a:
            i += 0.001
        i = i/0.4
        if list(str(i))[0] == 8:
            a = 'an'
        else:
            a = 'a'
        improve = f" {name.title()} could have received an A"
        improve += f" in the class if {pronouns[0]} scored {a}"
        improve += f" {str(round(i,2))}% on the final."
    # The appropriate grade according to the syllabus is defined for the
    # student's score
    if finale >= 89.5:
        returned = f"{name.title()}'s grade for the semester is an A."
        returned += f" I look forward to {pronouns[num]} continued success"
        returned += " next semester as we begin to explore more advanced"
        returned += " computer science and data science topics."
        return returned, finale
    elif 79.5 <= finale < 89.5:
        returned = f"{name.title()}’s grade for the semester is a B."
        returned += improve
        returned += " I look forward to a little more effort next semester"
        returned += " as we begin to explore more advanced"
        returned += " computer science and data science topics."
        return returned, finale
    elif 69.5 <= finale < 79.5:
        returned = f"{name.title()}’s grade for the semester is a C."
        returned += improve
        returned += " I hope to see improvements next semester as we"
        returned += " begin to explore more advanced computer science"
        returned += " and data science topics."
        return returned, finale
    elif 59.5 <= finale < 69.5:
        returned = f"{name.title()}’s grade for the semester is a D."
        returned += improve
        returned += " I hope to see major improvements next semester"
        returned += f" and I reccommend that {name} retakes the course."
        return returned, finale
    elif finale < 59.5:
        returned = f"{name.title()}’s grade for the semester is a F."
        returned += improve
        returned += f" {name.title()} needs to retake the class next semester"
        returned += " because of the failure."
        return returned, finale


def main():
    # If the teacher named their CSV file something different
    # after they recieved the copy that would be sent to them,
    # they can enter the name of their new file
    file = input('What is the name of your spreadsheet file? ')
    file += '.csv'
    # Creates header and student data
    head_studentz = head_students(file)
    # Below is the copy-and-paste semester topics that goes at the
    # beginning of every comment
    opening = "In the first semester, we covered the fundamentals"
    opening += " of Python, including loops, conditionals, functions,"
    opening += " and dictionaries. In addition, we explored connections"
    opening += " between math and computer science, including plotting,"
    opening += " simulating the game of craps, and optimizing customer"
    opening += " service at a bank."
    # The for loop below creates all the proper
    # information so that the functions can recieve the student data properly
    for student in head_studentz[-1]:
        name = student[head_studentz[0].index('first')]
        pronouns = student[head_studentz[0].index('pronouns (he/her/they)')]
        # Because the teacher only has to input the first
        # pronoun of the student, the code below creates the
        # rest of the pronouns
        if 'she' in pronouns.lower():
            pronouns = ['she', 'her', 'hers']
        elif 'they' in pronouns.lower():
            pronouns = ['they', 'them', 'their']
        else:
            pronouns = ['he', 'him', 'his']
        quiz = int(student[head_studentz[0].index('quiz')])
        problem = int(student[head_studentz[0].index('problem')])
        final = int(student[head_studentz[0].index('final')])
        bank = int(student[head_studentz[0].index('bank')])
        stack = student[head_studentz[0].index('used StackOverflow? (yes/no)')]
        hlp = student[head_studentz[0].index('asked for extra help? (yes/no)')]
        daily = student[head_studentz[0].index('worked alone daily? (yes/no)')]
        reflection = student[head_studentz[0].index('reflection')]
        opening2 = f"In {name.title()}'s self reflection,"
        opening2 += f" {pronouns[0]} wrote: "
        # The list below is the complete comment
        # All functions are referenced inside the list
        comment = [opening, '\n\n',
                   opening2, '\n\n',
                   reflection, '\n\n',
                   first(quiz,
                         problem,
                         final,
                         name,
                         pronouns,
                         finalgrade(final,
                                    problem,
                                    bank,
                                    quiz,
                                    name,
                                    pronouns)[-1]), '\n\n',
                   second(bank,
                          name,
                          pronouns), '\n\n',
                   third(daily,
                         stack,
                         hlp,
                         finalgrade(final,
                                    problem,
                                    bank,
                                    quiz,
                                    name,
                                    pronouns)[-1],
                         pronouns,
                         name), '\n\n',
                   finalgrade(final,
                              problem,
                              bank,
                              quiz,
                              name,
                              pronouns)[0]]
        name1 = student[head_studentz[0].index('first')] + ' '
        name1 += student[head_studentz[0].index('last')] + '.txt'
        with open(name1, 'w') as f:
            for item in comment:
                f.write(item)
    print("The comments have been written.")


if __name__ == '__main__':
    main()
```
